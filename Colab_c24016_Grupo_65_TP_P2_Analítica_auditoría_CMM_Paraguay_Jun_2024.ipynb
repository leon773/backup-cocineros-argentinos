# Normalización

```python
#Se normaliza la primer columna para todas tablas, llamandose id_distributor y siendo del mismo tipo
df_distributors['id'] = df_distributors['id'].str.replace('.','')
df_distributors['id_distributor'] = df_distributors['id']
df_distributors.drop(columns=['id'],inplace = True)
df_distributors = df_distributors[['id_distributor'] + [col for col in df_distributors.columns if col != 'id_distributor']]
df_distributors=df_distributors.astype({'id_distributor': 'int64'})
df_distributors = df_distributors.sort_values(by='id_distributor')
df_distributors = df_distributors.reset_index(drop=True)
df_distributors
```

```python
df_exports['distributor'] = df_exports['distributor'].str.replace('.','')
df_exports['id_distributor'] = df_exports['distributor']
df_exports.drop(columns=['distributor'],inplace = True)
df_exports = df_exports[['id_distributor'] + [col for col in df_exports.columns if col != 'id_distributor']]
df_exports=df_exports.astype({'id_distributor': 'int64'})
df_exports = df_exports.sort_values(by='id_distributor')
df_exports = df_exports.reset_index(drop=True)
df_exports
```

```python
df_sales=df_sales.astype({'distributor': 'int64'})
df_sales['id_distributor'] = df_sales['distributor']
df_sales.drop(columns=['distributor'],inplace = True)
df_sales = df_sales[['id_distributor'] + [col for col in df_sales.columns if col != 'id_distributor']]
df_sales = df_sales.sort_values(by='id_distributor')
df_sales = df_sales.reset_index(drop=True)
df_sales
```

# Modelado

Una tabla util podria ser la diferencia entre exports y sales. Ya que mostraria si se gano o perdio dinero en cada distribuidor segun el producto.

```python
df_diferencia = df_sales.set_index('id_distributor').subtract(
    df_exports.set_index('id_distributor'), fill_value=0).reset_index()
dif_columns = df_diferencia.columns
ndif_names = {col: col + '_dif' for col in dif_columns[1:]}
df_diferencia.rename(columns=ndif_names, inplace=True)
df_diferencia
```

Se unifican todas las tablas en una misma, salvo locations que esta incompleta y no parece tener un uso real



```python
df_all = pd.merge(df_distributors, df_exports, on='id_distributor')
df_all = pd.merge(df_all, df_sales, on='id_distributor')
df_all = pd.merge(df_all, df_diferencia, on='id_distributor')
df_all.columns = [col.replace('_x', '_exports').replace('_y', '_sales') for col in df_all.columns]
df_all
```

```python
#Ver los numeros negativos en rojo pueden ayudar mucho a identificar rapidamente donde se vendio muy poco o nada segun el producto.
df_styled_diferencia = df_diferencia.style.applymap(color_negative_red, subset=pd.IndexSlice[:, df_diferencia.columns[1:]])
df_styled_diferencia = df_styled_diferencia.format(format_two_decimals, subset=pd.IndexSlice[:, df_diferencia.columns[1:]])

# Mostrar el DataFrame resultante con el estilo aplicado
df_styled_diferencia
```

```python


# Crear el DataFrame de porcentaje
df_porcentaje = df_diferencia.copy()

# Calcular el porcentaje para todas las columnas excepto la primera
for col in df_porcentaje.columns[1:]:
    df_porcentaje[col] = ((df_diferencia[col]) / df_exports[col.replace('_dif', '')]) * 100

# Aplicar el estilo y formato al DataFrame
styled_df_porcentaje = df_porcentaje.style.applymap(color_negative_red, subset=pd.IndexSlice[:, df_porcentaje.columns[1:]])
styled_df_porcentaje = styled_df_porcentaje.format(format_percentage, subset=pd.IndexSlice[:, df_porcentaje.columns[1:]])

# Mostrar el DataFrame resultante con el estilo aplicado
styled_df_porcentaje
```

Verlo como porcentaje sobre el total de lo importado ayuda a identificar mas facilmente cuanta ganancia hubo o de lo contrario perdida. Hay que tener en cuenta que en estos casos el % negativo indica que ni siquiera se llego a recuperar el costo de la inversion.
Si bien los materiales no son pedecederos por lo que no es una perdida directa, tener mercaderia parada implica una perdida en costos de almacenaje y perdida de oportunidad de haber utilizado esos recursos en otro sitio que si genere ganancia. Tambien se puede tener en cuenta que los ladrillos(Brick) y el acero(steel) pueden haber tenido ventas nulas ya que habia distribuidores que los fabricaban de forma local, lo cual puede ser mas economico que importarlos.

```python
df_sales_total=df_sales
df_sales_total['Ventas total'] = df_sales.iloc[:, 1:].sum(axis=1)
df_sales_total = df_sales[['id_distributor', 'Ventas total']]
df_sales_total
```

```python
df_exports_total=df_exports
df_exports_total['Exportaciones totales'] = df_exports.iloc[:, 1:].sum(axis=1)
df_exports_total = df_exports[['id_distributor', 'Exportaciones totales']]
df_exports_total
```

```python

df_diferencia_total=df_diferencia
df_diferencia_total['Diferencia total'] = df_diferencia.iloc[:, 1:].sum(axis=1)
df_diferencia_total = df_diferencia[['id_distributor', 'Diferencia total']]
df_diferencia_total
```

```python
# Crear el DataFrame de porcentaje
df_porc_total = df_diferencia_total.copy()

# Calcular el porcentaje para la columna 'total'
df_porc_total['% Total'] = ((df_diferencia_total['Diferencia total']) / df_exports_total['Exportaciones totales']) * 100
df_porc_total = df_porc_total[['id_distributor', '% Total']]
df_porc_total
```

```python
# Quedarse solo con las columnas 'distributor', 'Steel_dif' y 'Plexiglass_dif'
df_mayorperdida_acum = df_mayorperdida[['distributor', 'Steel_dif', 'Plexiglass_dif']]
# Llevar Steel_dif de Collins LLC y Mraz Group a 0
df_mayorperdida_acum.loc[df_mayorperdida_acum['distributor'].isin(['Collins LLC', 'Mraz Group']), 'Steel_dif'] = 0

# Llevar Plexiglass_dif de Brekke-Stiedemann a 0
df_mayorperdida_acum.loc[df_mayorperdida_acum['distributor'] == 'Brekke-Stiedemann', 'Plexiglass_dif'] = 0
# Calcula la suma de cada columna (excepto "distributor")
total_por_columna = df_mayorperdida_acum.drop(columns=['distributor']).sum()

# Agrega una nueva fila con los totales
df_mayorperdida_acum.loc['Total'] = total_por_columna

# Calcula la suma de cada fila (excepto "distributor")
total_por_fila = df_mayorperdida_acum.drop(columns=['distributor']).sum(axis=1)

# Agrega una nueva columna con los totales
df_mayorperdida_acum['Total'] = total_por_fila
# Reemplaza NaN por "Total" en la columna "distributor" en el índice "Total"
df_mayorperdida_acum.loc['Total', 'distributor'] = 'Total'


df_mayorperdida_acum
```
